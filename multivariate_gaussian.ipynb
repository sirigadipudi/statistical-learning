```python
import numpy as np
import matplotlib.pyplot as plt

# This is similar to the code from lecture 1
# to sample from a 2D Gaussian Distribution
mean = [0,0]
cov = [[1, 1.5], [1.5, 5]]
# Fix random seed to get consistent results
np.random.seed(1024)
X = np.random.multivariate_normal(mean, cov, 1000)
fig, ax = plt.subplots(figsize=(6, 6))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
# Set x/y axis limits
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
plt.xlabel('x')
plt.ylabel('y')
plt.title('2D scatter plot of a multivariate Gaussian distribution')
plt.show()
#fig
```

```python
# TODO: Report the estimate mean and covariance
# of the sampled points
mean_X = sum(X)/len(X)
print("Mean: ",mean_X)
cov_X = np.cov(X.T)
print("Covariance: ",cov_X)
#print(cov_X.shape)
```

```python
# TODO: Plot the histogram for the x-coordinates of X
# and y-coordinates of X respectively.
# You can use the plt.hist() function

fig1 = plt.figure(figsize=(6,6))
fig1 = plt.hist(X[:,0],bins=100,density=False,color='r',alpha=0.5,label='x')
#fig1 = plt.hist(X[:,1],bins=100,density=False,color='g',alpha=0.5,label='y')
plt.legend(loc='upper right')
plt.title('Histogram plot')
plt.show()
#fig1
```

```python
# TODO: Are the x-coordinates of X samples from
# some Gaussian distribution?
# If so, estimate the mean and variance.
# Do the same for the y-coordinates.
mean_Xx = np.mean(X[:,0])
print('Mean of x-coordinates: ',mean_Xx)
variance_Xx = np.var(X[:,0])
print('Variance of x-coordinates: ',variance_Xx)

mean_Xy = np.mean(X[:,1])
print('Mean of y-coordinates: ',mean_Xy)
variance_Xy = np.var(X[:,1])
print('Variance of y-coordinates: ',variance_Xy)


# TODO: Generate a new 2D scatter plot of 1000 points,
# such that the x-coordinates(y-coordinates) of all the points
# are samples from a 1D Gaussian distribution 
# using the estimated mean and variance based on the x-coordinates(y-coordinates) of X.

x_gaussian = np.random.normal(mean_Xx, variance_Xx, 1000)
y_gaussian = np.random.normal(mean_Xy, variance_Xy, 1000)
plt.scatter(x_gaussian, y_gaussian, c='g', s=10.0, alpha=0.3, label="2D-Gaussian X1")
plt.title('2-D Scatter plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Back to the original X
fig, ax = plt.subplots(figsize=(6, 6))
# c='r', dot color is red
# s=10.0, dot size is 10
# alpha=0.3, dot opacity is 0.3
ax.scatter(X[:,0], X[:,1], c='r', s=10.0, alpha=0.3, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])

# TODO: Plot a line segment with x = [-10, 10]
# and y = 3x + 1 onto the 2D-Gaussian plot.
# The np.linspace() function may be helpful.

x = np.linspace(-10,10,10)
y = 3*x + 1
plt.plot(x,y,color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# TODO: Project X onto line y=3x + 1
# and plot the projected points on the 2D space.
# You need to remove this line and assign the projected points to X_proj.
#X_proj = np.zeros(X.shape)

# Xp = (X[:,0] + 3*X[:,1] - 3)/10
# Yp = (X[:,0] + 9*X[:,1] + 1)/10

# X_proj = np.array([[Xp,Yp]])

# Define line y = 3x + 1
line_vector = np.array([3, -1])

# Compute projections onto line
projections = (np.dot(X, line_vector) + 1) / np.dot(line_vector, line_vector)
y1 = np.reshape(projections, (1000,1))
X_proj = X - y1*line_vector

#Here we only plot 30 points to check the correctness

fig, ax = plt.subplots(figsize=(6.5, 6.5))
ax.scatter(X_proj[:,0], X_proj[:,1], c='b', s=10.0, alpha=1.0, label="Projected 2D-Gaussian")
ax.scatter(X[:30,0], X[:30,1], c='r', s=10.0, alpha=1.0, label="2D-Gaussian")
ax.grid()
ax.legend(loc = 0)
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
plt.plot(x,y,color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

#You can also add the line from the previous plot for verification.
```
